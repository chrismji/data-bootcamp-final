The variables that I have decided to keep and use for the models are its weight, BST, invidual stats, typing, and generation. Stat totals and typing should make sense, as those are the core values that define a Pokemon in gameplay. And as mentioned before, generation data is being included in order to try to determine 'power-creep' is a legitimate concept in competitive Pokemon.

The most suprising addition here is probably weight, especially since I am excluding height despite both being 'physical' traits. While these numbers typically have zero gameplay impact, there are multiple commonly-seen moves that actually depend on a Pokemon's weight for damage calculations (heavier pokemon = more damage). As such, while it may be weak, weight could possibly have a tangible impact on a Pokemon's viability.

Notably, I am not including Pokemon's abilities nor am I taking into account movesets. While these traits are integral to a Pokemon's competitive identity (even moreso than numerical stats), it's simply too difficult to categorize. There are several hundred unique abilities, even more movesets, and too many moves/abilities are 'signature' ones unique to only a handful of Pokemon each. Using a model to analyze a Pokemon's abilities or learnset is simply out of the scope of a project like this.
